## 11. Equação da onda e diferenças finitas


([Boyce e Diprima, 2024](https://integrada.minhabiblioteca.com.br/reader/books/9788521638841/epubcfi/6/44%5B%3Bvnd.vst.idref%3Dchapter10%5D!/4/724%5Bsec10-7%5D/3:38%5BCor%2Cda%20%5D)) Umaoutra equação diferencial parcial que ocorre com frequência em Matemática aplicada é a equação de onda. Alguma forma dessa equação, ou uma de suas generalizações, aparece quase que inevitavelmente em qualquer análise matemática de fenômenos envolvendo a propagação de ondas em um meio contínuo. Por exemplo, todos os estudos de ondas acústicas, ondas de água, ondas eletromagnéticas e ondas sísmicas baseiam-se nesta equação.

Suponha que uma corda elástica de comprimento $L$ esteja ligeiramente esticada entre dois suportes no mesmo nível horizontal, de modo que o eixo dos $x$ esteja ao longo da corda


<img src="https://raw.githubusercontent.com/tiagoburiol/MatComp_IV/refs/heads/main/imagens/corda.png" width="360"/>



no domínio $0<x<L, t>0$. Esta equação é conhecida como equação de onda unidimensional.

Suponha também  que a corda é colocada em movimento (sendo puxada, por exemplo) de modo que vibra em um plano vertical, e denote por $u(x, t)$ o deslocamento vertical da corda no ponto $x$ no instante $t$. Se forem desprezados os efeitos de amortecimento, como a resistência do ar, e se a amplitude do movimento não for muito grande, então $u(x, t)$ satisfará a equação diferencial parcial

$$
a^2 u_{x x}=u_{t t}$$


no domínio $0<x<L, t>0$. Esta equação é conhecida como **equação da onda unidimensional**. 

O coeficiente constante $a^2$ que aparece é dado por

$$
a^2=\frac{T}{\rho},
$$

em que $T$ é a tensão (força) na corda e $\rho$ é a massa por unidade de comprimento do material da corda. Assim, $a$ é a velocidade de propagação das ondas ao longo da corda.

Para descrever completamente o movimento da corda, é necessário especificar, também, condições iniciais (posição inicial e velocidade inicial) e de condições de contorno

$$
u(x, 0)=f(x), \quad u_t(x, 0)=g(x), \quad 0 \leq x \leq L
$$

e

$$
u(0, t)=0, \quad u(L, t)=0, \quad t \geq 0
$$

Também é necessário supor que

$$
f(0)=f(L)=0, \quad g(0)=g(L)=0.
$$

A equação de onda a duas dimensões é

$$
a^2\left(u_{x x}+u_{y y}\right)=u_{t t} .
$$

Analogamente, em três dimensões, a equação de ondas é

$$
a^2\left(u_{x x}+u_{y y}+u_{z z}\right)=u_{t t} .
$$

#### Corda Elástica com Deslocamento Inicial Não Nulo: Solução Analítica

Primeiro, suponha que a corda é deslocada em relação a sua posição de equilíbrio e depois solta, no instante $t = 0$, com velocidade nula para vibrar livremente. Então, o deslocamento vertical $u(x, t)$ tem de satisfazer a equação 

$$
a^2 u_{x x}=u_{t t}, \quad 0<x<L, \quad t>0
$$

as condições de contorno

$$
u(0, t)=0, \quad u(L, t)=0, \quad t \geq 0
$$

e as condições iniciais

$$
u(x, 0)=f(x), \quad u_t(x, 0)=0, \quad 0 \leq x \leq L.
$$

Pelo método de separação de variáveis escrevemos

$$
u(x, t)=X(x) T(t)
$$

e, substituindo na equação da onda, obtemos

$$
\frac{X^{\prime \prime}}{X}=\frac{1}{a^2} \frac{T^{\prime \prime}}{T}=-\lambda,
$$

em que $\lambda$ é uma constante de separação. Vemos, então, que $X(x)$ e $T(t)$ satisfazem as equações diferenciais ordinárias

$$
X^{\prime \prime}+\lambda X=0 \quad \text{e} \quad T^{\prime \prime}+a^2 \lambda T=0 .
$$

Substituindo $u(x, t)$ nas condições de contorno, vemos que $X(x)$ tem de satisfazer $X(0)=0$ e  $X(L)=0$. Vemos também que $T(t)$ tem de satisfazer a condição inicial $T^{\prime}(0)=0$.

O problema é precisamente o mesmo que foi resolvido para a equação do calor, já conhecido, e terá soluções não triviais se, e somente se, $\lambda$ for um autovalor,

$$
\lambda=\frac{n^2 \pi^2}{L^2}, \quad n=1,2, \ldots,
$$

e $X(x)$ for proporcional à autofunção correspondente $\operatorname{sen}(n \pi x / L)$.
Usando os valores de $\lambda$ na equação em $T$, obtemos

$$
T^{\prime \prime}+\frac{n^2 \pi^2 a^2}{L^2} T=0
$$

Portanto,

$$
T(t)=k_1 \cos \left(\frac{n \pi a t}{L}\right)+k_2 \operatorname{sen}\left(\frac{n \pi a t}{L}\right),
$$

em que $k_1$ e $k_2$ são constantes arbitrárias. A condição inicial $T^{\prime}(0)=0$ implica que $k_2=0$, de modo que $T(t)$ tem de ser proporcional a $\cos (n \pi a t / L)$.

Assim, as funções

$$
u_n(x, t)=\operatorname{sen}\left(\frac{n \pi x}{L}\right) \cos \left(\frac{n \pi a t}{L}\right), \quad n=1,2, \ldots
$$

são as soluções fundamentais do problema dado.

Para satisfazer a condição inicial (não homogênea) que falta, vamos considerar uma superposição das soluções fundamentais com coeficientes escolhidos adequadamente. Supomos, então, que $u(x, t)$ tem a forma

$$
u(x, t)=\sum_{n=1}^{\infty} c_n u_n(x, t)=\sum_{n=1}^{\infty} c_n \operatorname{sen}\left(\frac{n \pi x}{L}\right) \cos \left(\frac{n \pi a t}{L}\right),
$$

em que as constantes $c_n$ ainda têm de ser escolhidas. A condição inicial $u(x, 0)=f(x)$ implica que

$$
u(x, 0)=\sum_{n=1}^{\infty} c_n \operatorname{sen}\left(\frac{n \pi x}{L}\right)=f(x)
$$

Em consequência, os coeficientes $c_n$ têm de ser os coeficientes na série de Fourier em senos de $f$ com período $2 L ; \operatorname{logo}$,

$$
c_n=\frac{2}{L} \int_0^L f(x) \operatorname{sen}\left(\frac{n \pi x}{L}\right) d x, \quad n=1,2, \ldots
$$

**Exemplo 1:**([Boyce e Diprima, 2024](https://integrada.minhabiblioteca.com.br/reader/books/9788521638841/epubcfi/6/44%5B%3Bvnd.vst.idref%3Dchapter10%5D!/4/854/4/8%5Bfig10-7-3%5D/4/3:54%5Bma%20%2Ccor%5D)) Considere uma corda vibrante de comprimento $L=30$ que satisfaz a equação de onda

$$
4 u_{x x}=u_{t t}, \quad 0<x<30, \quad t>0 .
$$

Suponha que as extremidades da corda estão fixas e que a corda é colocada em movimento sem velocidade inicial a partir da posição inicial

$$
u(x, 0)=f(x)=\left\{\begin{array}{lr}
x / 10, & 0 \leq x \leq 10 \\
(30-x) / 20, & 10<x \leq 30
\end{array}\right.
$$

Vamos encontrar o deslocamento $u(x, t)$ da corda e seu movimento durante um período por meio de alguns gráficos.

```python
import sympy as sp
x = sp.symbols('x')
n = sp.symbols('n', integer=True, positive=True)
```

```python
a = 2
L = 30
```

```python
c = (2/L)*(sp.integrate((x/10)*sp.sin(n*sp.pi*x/L),(x,0,10)) +\
       sp.integrate(((30-x)/20)*sp.sin(n*sp.pi*x/L),(x,10,30)))

sp.simplify(c, trig =True)
```

Assim, usando SymPy obtemos os coeficientes
$$
c_n=\frac{9}{n^2 \pi^2} \operatorname{sen}\left(\frac{n \pi}{3}\right), \quad n=1,2, \ldots
$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [4,2]
```

```python
#def cn(n):
#    coef = (9/(n**2*np.pi**2))*np.sin(n*np.pi/3)
#    return coef

cn = sp.lambdify(n,c)
```

```python
a = 2
L = 30

def u(x, t, n_max): #aqui, x é um array, t é fixo
    soma = 0
    for n in range(1,n_max+1):
        soma += cn(n)*np.sin(n*np.pi*x/L)*np.cos(a*n*np.pi*t/L)
    return soma
```

```python
x = np.linspace(0,30,30)

us = []
for t in np.linspace(0,60,50):
    us.append(u(x,t,50))   
    plt.plot(x,u(x,t,50), color='b', lw=0.5)
us = np.array(us)
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots()
line, = ax.plot([], [])
ax.set_xlim(0, us.shape[1]-1)
ax.set_ylim(us.min(), us.max())

def update(frame):
    line.set_data(x, us[frame])
    return line,

ani = FuncAnimation(fig, update, frames=us.shape[0], 
                    interval=50, blit=False)
```

#### Diferenças finitas

([Burden, 2016](https://integrada.minhabiblioteca.com.br/reader/books/9788522123414/pageid/847)) Agora, consideramos a resolução numérica da equação da onda, um exemplo de uma equação diferencial parcial hiperbólica. A equação da onda é dada pela equação diferencial

$$
\frac{\partial^2 u}{\partial t^2}(x, t)-\alpha^2 \frac{\partial^2 u}{\partial x^2}(x, t)=0, \quad 0<x<l, \quad t>0
$$

sujeita às condições

$$
u(0, t)=u(l, t)=0, \quad \text { para } t>0 {, }
$$

$$
u(x, 0)=f(x) \quad \text { e } \quad \frac{\partial u}{\partial t}(x, 0)=g(x), \quad \text { para } 0 \leq x \leq l \text {, }
$$

em que a é uma constante que depende das condições físicas do problema.

Em qualquer ponto interior da malha, $(x_i, t_j)$, a equação da onda torna-se

$$
\frac{\partial^2 u}{\partial t^2}\left(x_i, t_j\right)-\alpha^2 \frac{\partial^2 u}{\partial x^2}\left(x_i, t_j\right)=0 {.}
$$

Utilizando o quociente das diferenças centradas para as segundas derivadas parciais, a equação de diferença (onde $w$ é a aproximação para $u$) tem a forma:

$$
\frac{w_{i, j+1}-2 w_{i, j}+w_{i, j-1}}{k^2}-\alpha^2 \frac{w_{i+1, j}-2 w_{i, j}+w_{i-1, j}}{h^2}=0 .
$$

Fazendo $\lambda=\alpha k / h$ podemos escrever a equação de diferença como

$$
w_{i, j+1}-2 w_{i, j}+w_{i, j-1}-\lambda^2 w_{i+1, j}+2 \lambda^2 w_{i, j}-\lambda^2 w_{i-1, j}=0
$$

isolando $w_{i,j+1}$ obtemos

$$
w_{i, j+1}=2\left(1-\lambda^2\right) w_{i, j}+\lambda^2\left(w_{i+1, j}+w_{i-1, j}\right)-w_{i, j-1} .
$$

Esta equação vale para cada $i= 1, 2,..., m-1$ e $j=1, 2,...$. As condições de contorno fornecem

$$
w_{0, j}=w_{m, j}=0, \text { para cada } j=1,2,3, \ldots
$$

e a condição inicial implica que 

$$
w_{i, 0}=f\left(x_i\right) \text {, para cada } i=1,2, \ldots, m-1 .
$$

Escrevendo este conjunto de equações em forma matricial, tem-se

<img src="https://raw.githubusercontent.com/tiagoburiol/MatComp_IV/refs/heads/main/imagens/sistema_onda.png" width="600"/>

As equações acima implicam que o ( $j+1$ )-ésimo passo no tempo necessita dos valores dos $j$-ésimo e ( $j-1$ )-ésimo passos no tempo. Isto produz um pequeno problema de partida, já que os valores para $j=1$, que são necessários na para calcular $w_{i, 2}$, devem ser obtidos da condição de velocidade inicial

$$
\frac{\partial u}{\partial t}(x, 0)=g(x), \quad 0 \leq x \leq l .
$$

Em [Burden e Faires (2016)](https://integrada.minhabiblioteca.com.br/reader/books/9788522123414/pageid/846) é mostrada uma abordagem na qual é usada a seguinte equação de diferença para encontrar $w_{i,1}$, para cada $i=1,2,...,m-1$

$$
w_{i, 1}=\left(1-\lambda^2\right) f\left(x_i\right)+\frac{\lambda^2}{2} f\left(x_{i+1}\right)+\frac{\lambda^2}{2} f\left(x_{i-1}\right)+k g\left(x_i\right).
$$



 É suposto que exista um limitante superior $T$ para o valor de $t$, que será usado na técnica de parada e que $k = T/N$, em que $N$ também é dado.

Em [Nogueira (2025)](https://github.com/JonathanWNogueira/burden_numerical_analysis/blob/main/12.ipynb) é mostrado uma implementaçao do algoritmo disponível em [Burden e Faires (2016)](https://integrada.minhabiblioteca.com.br/reader/books/9788522123414/pageid/849) para obter uma aproximação para a solução da equação da onda por diferenças finitas. Uma versão "vetorizada" desse  algoritmo é mostrada no exemplo a seguir.  

**Exemplo 2:** ([Burden e Faires, 2016](https://integrada.minhabiblioteca.com.br/reader/books/9788522123414/pageid/850)) Determine uma aproximação para a solução do problema hiperbólico

$$
\frac{\partial^2 u}{\partial t^2}(x, t)-4 \frac{\partial^2 u}{\partial x^2}(x, t)=0, \quad 0<x<1, \quad 0<t,
$$

com as condições de contorno

$$
u(0, t)=u(1, t)=0, \quad \text { para } \quad 0<t
$$

e condições iniciais

$$
u(x, 0)=\operatorname{sen}(\pi x), \quad 0 \leq x \leq 1, \quad \text { e } \quad \frac{\partial u}{\partial t}(x, 0)=0, \quad 0 \leq x \leq 1,
$$

usando $h=0,1$ e $k=0,05$. Compare os resultados com a solução exata.

$$
u(x, t)=\operatorname{sen} \pi x \cos 2 \pi t
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l, T = 1, 1
m, N = 10, 20
alpha = 2
a, b = 0, 0
```

```python
h = l/m
k = T/N
lamb = k*alpha/h

x = np.linspace(0,l,m+1)
f0 = np.sin(np.pi*x)
df0 = np.zeros_like(x)

w = np.zeros((N+1, m+1))
w[0] = f0

# passo n = 1
w[1,1:-1] = ((1 - lamb**2)*f0[1:-1] +
             0.5*lamb**2*(f0[2:] + f0[:-2]) +
             k*df0[1:-1])

# passos n ≥ 2
for n in range(2, N+1):
    w[n,1:-1] = (2*(1 - lamb**2)*w[n-1,1:-1] +
                 lamb**2*(w[n-1,2:] + w[n-1,:-2]) -
                 w[n-2,1:-1])
```

```python
print(w[20,:])
```

```python
fig, ax = plt.subplots()
line, = ax.plot([], [])
ax.set_xlim(0, w.shape[1]-1)
ax.set_ylim(w.min(), w.max())

def update(frame):
    line.set_data(x/h, w[frame])
    return line,

ani = FuncAnimation(fig, update, frames=w.shape[0], 
                    interval=50, blit=False)
```
