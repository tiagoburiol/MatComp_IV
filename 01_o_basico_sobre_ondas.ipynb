# 1. O básico sobre ondas

Na física, uma onda é uma perturbação que viaja pelo espaço e pela matéria com a transferência de energia de um lugar para outro. Modelos matemáticos que descrevem ondas são usados por engenheiros para tratar de sistemas que oscilam ou vibram. Neste contexto de problemas, as funções trigonométricas desempenham um papel fundamental. 

Para entender como as elas se formam, se propagam e suas características precisamos primeiro nos familiarizar com o modelo matemático de uma onda. Podemos descrever uma onda variando tanto no tempo quanto no espaço. Para uma única onda tem-se algumas características a serem observadas.

**Amplitude:** medida da magnitude máxima da perturbação causada pela onda em relação à sua posição de equilíbrio ou valor médio.

**Período:** intervalo de tempo (ou da variável independente) necessário para que um ciclo completo da onda se repita.

**Frequência angular:** está relacionada à velocidade com que uma onda ou um sistema oscilante completa um ciclo em radianos por unidade de tempo.

**Frequência comun:** nos diz quantos ciclos completos ocorrem por unidade de tempo.

Funções periódicas (ou sinais periódicos) frequentemente são tratados como ondas. Um sinal periódico é definido como aquele que apresenta repetição de seus valores a um intervalo definido de tempo (período):

$$f(\theta) = f(\theta + nT)$$

onde $T$ é o valor deste intervalo (período) e $n$ indica um número inteiro.


Considerando o ângulo $\theta$ como uma função do tempo, tal que $\theta = \omega t$, para
um intervalo $\Delta \theta$ temos $\Delta \theta = \omega \Delta t$ ou $\Delta t = \frac{\Delta \theta}{\omega}$ onde:
- $t=$ tempo ($s$)
- $\theta=$ ângulo ($rad$)
- $\omega=$ freq.angular ($rad/s$)

Assim, para a função completar um ciclo (2$\pi$), o tempo associado é definido como **período** 

$$T = \frac{\Delta \theta}{\omega} = \frac{2\pi}{\omega}$$ 

em segundos ($s$).

O inverso do período, representando a quantidade de ciclos por unidade de tempo, é denominado **frequência**: 

$$f=\frac{1}{T} = \frac{\omega}{2\pi}$$

em ciclos por segundo, ou hertz ($Hz$).

Uma forma ondulatória geral descrita por por uma função seno ou cosseno, denominada senóide pode ser escrita como uma função na forma

$$ f(t) = C+ A sen(\omega t + \phi)$$

onde $\phi$ é o chamado ângulo de fase, $A$ é a amplitude, $\omega$ é frequência angular e $C$ é uma tranlação vertical (ou valor médio, determina a altura média acima da abscissa). O período é dado por $T = \frac{2\pi}{\omega}$.

Não existe nenhuma convenção bem estabelecida para escolher qualquer uma das duas funções seno ou cosseno, de qualquer modo, os resultados seriam idênticos.


**Exemplo 1.** O código abaixo mostra a construção da figura de uma senóide, mostrando o período e a amplitude.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 2.0 
T = 3.0
phi = 0.5
t = np.linspace(-2*np.pi, 2*np.pi, 1000)

w = 2*np.pi/T
y = A*np.sin(w*t+phi)

fig, ax = plt.subplots(figsize=(12,2))

#desenhando a medida do periodo (comprimento)
ax.text(T/2,A+0.6, "T", style='italic', size='x-large')
ax.hlines(y= A+0.4, xmin=0, xmax= T, color='k', lw=1.0)
ax.plot(0, A+0.4, 'k|')
ax.plot(T, A+0.4, 'k|')

# desenhando a medida da amplitude
ax.text(-0.3,A/2, "$A$", style='italic', size='x-large')
ax.vlines(x= 0, ymin=0, ymax= A, color='k', lw=1.0)
ax.plot(0, 0, 'k_')
ax.plot(0, A, 'k_')


# marcando alguns valores uteis, míltiplos de pi
ax.text(np.pi/2, -(A+2.5), "$\\frac{\\pi}{2}$", size='x-large')
ax.vlines(np.pi/2, -(A+1.5), A+1.5, linestyles=[(0,(1,1))])
ax.text(np.pi, -(A+2.5), "$\\pi$", size='x-large')
ax.vlines(np.pi, -(A+1.5), A+1.5, linestyles=[(0,(1,1))])


ax.set_ylim(-(A+1.5),A+1.5)         
plt.plot(t,y)
plt.grid()
```

A curva da figura acima é dita descrita no *domínio do tempo*. Embora não seja tão familiar, o *dominio de frequência* fornece uma perspectica alternativa para caracterizar o comportamento de funções oscilantes. Conhecendo a amplitude máxima positiva de uma função senoidal, a posição de $1/T$ no *eixo de frequências* e o ângulo de fase, tem-se informações suficientes para reproduzir a curva no domínio do tempo. A descrição de sinais no domínio da frequência é usada para caracterizar sinais mais complicadas usando a transformada de Fourier.

Vamos lembrar agora, que algumas equações diferenciais tem soluções escritas em termos se seno e cosseno. Por exemplo, a equação $m y^{\prime \prime}+k y=0$, que descreve um modelo sistema mecânico massa-mola, tem solução $y(t)=A \cos \omega_0 t+B \operatorname{sen} \omega_0 t$, em que $\omega_0=\sqrt{k/m}$.

Como as funções trigonométricas na solução possuem o período $2\pi/ \omega_0$ , o corpo executa $\omega_0/2\pi$ ciclos por segundo (ou Hz) essa frequencia é chamada de **frequência natural**.

A função $y(t)=A \cos \omega_0 t+B \operatorname{sen} \omega_0 t$ pode ser combinada na forma de um cosseno envolvendo um desvio de fase, com uma amplitude $C=\sqrt{A^2+B^2}$ e um ângulo de fase $\delta=\arctan (B / A)$,

$$y(t)=C \cos \left(\omega_0 t-\delta\right).$$

### A onda quadrada

Para consruir uma onda quadrada em python é conveniente utilizar uma sentença condicional que retorna $1$ (`True`) se for verdadeira ou $0$ se for falsa(`False`). Por exemplo, para avaliar quais valores inteiros entre $-5$ e $5$ são maiores que $1$ podemos fazer como é mostrado a seguir:

```python
x =  np.arange(-5,5)
print('Inteiros x entre -5 e 5:', x)
print('Condicional maior que 1 :', x > 1 )
print('Valor correspondente:',np.int32(x > 1))
```

Vejamos como usar uma  sentença condicional para construir uma onda quadrada  no exemplo a seguir.

**Exemplo 3:** Para construir uma única onda quadrada definida pela função

$$f(t) = \begin {cases} 
                 1, \,\, |t|<1   \\ 
                 0, \,\, |t|\geq1
                 \end{cases}$$ 
no intervalo $-2\leq t \leq2$, podemos fazer como é mostrado a abaixo:

```python

t = np.linspace(-2,2,1000)

fig, ax = plt.subplots(figsize=(4,2))
plt.plot(t, (t>-1)&(t<1))
plt.grid()
plt.show()
```

**Observação:** As linhas verticais em $x=-1$ e $x=1$ não fazem parte do gráfico da função. Uma representação mais precisa deveria omotir essas linhas e usar a simbologia de intervalo fechado e aberto nesses pontos. 

Para obter um sinal periódico repetindo a onda do exemplo anterior (de período $T=4$) ao longo de um intervalo maior, por exemplo $-6\leq t \leq 6$, podemos fazer como é mostrado no exemplo a seguir.

**Exemplo 4:** Um sinal periódico repetindo uma onda quadrada.

```python
def f(x):
    if abs(x)>2: 
        return f(abs(x)-4)
    if (x>0)&(x<1) or (x<0)&(x>-1): 
        return 1
    else:
        return 0
    
fig, ax = plt.subplots(figsize=(10,3))
x = np.linspace(-6,6,1000)
plt.plot(x,[f(xi) for xi in x])
plt.grid()
```

**Exemplo 5.** Gráfico da função $f(t)=2cos(2\pi t)$, de amplitude $A=2$ e período $T=1$ pode ser gerado como é mostrado abaixo.

```python
t = np.linspace(-4,4,1000)
f = lambda t: 2*np.cos(2*np.pi*t)

fig, ax = plt.subplots(figsize=(10,2))
plt.plot(t, f(t))
plt.grid()
```

A função `np.sign()` determina o sinal (positivo, negativo ou zero) dos elementos de um `array`. A função retorna um `array` com $1$ para os elementos positivos, $-1$ para os negativos e $0$ para os nulos. Assim, uma onda quadrada com período $T=1$ pode ser gerada a partir da função do exemplo anterior como é mostrado a seguir.

**Exemplo:** Gráfico da onda quadrada par, de amplitude $A=1$ e período $T=1$.

```python
fig, ax = plt.subplots(figsize=(10,2))
plt.plot(t, np.sign(f(t)))
plt.grid()
```

**Aproximação por séries trigonométricas**

Chamamos de *senoide* qualquer forma ondulatória que possa ser descrita por um seno ou um cosseno, por exemplo, usando cosseno podemos representar pela forma geral

$$f(t)=A_0+C_1 \cos \left(\omega_0 t+\theta\right)$$

Em que o *valor médio* $A_0$ determina a altura média acima da abscissa (pode ser interpretado como uma translação vertical). A *amplitude* $C_1$ especifica a altura da oscilação. A *frequência angular* $\omega_0$ caracteriza com que frequência o ciclo ocorre. Finalmente, o ângulo de fase, ou *deslocamento angular*, $\theta$ fornece a amplitude do deslocamento horizontal inicial da curva senoidal.

Observe que, usando a identidade trigonométrica para a soma de arcos, podemos escrever

$$
f(t)=A_0+C_1\Big[\cos \left(\omega_0 t\right) \cos (\theta)-\operatorname{sen}\left(\omega_0 t\right) \operatorname{sen}(\theta)\Big]
$$

**Exemplo:** Faça o gráfico da reta horiontal $f_1(t)=1,7$, e das senoides $f_2(t)=0.5\cos(2\pi t)$ e $f_3(t)=-0.866\operatorname{sen}(2\pi t)$. Então, faça o gráfico da função $f_4(t)=1,7+cos(2\pi t + \pi/3)$. Note que $f_1+f_2+f_3=f_4$.


```python
f1 = lambda t: np.full(len(t),1.7)
f2 = lambda t: 0.5*np.cos(2*np.pi*t)
f3 = lambda t: -0.866*np.sin(2*np.pi*t)
f4 = lambda t: 1.7+np.cos(2*np.pi*t+np.pi/3)

t = np.linspace(0,2)

plt.figure(figsize = (7, 2))
plt.plot(t,f1(t))
plt.plot(t,f2(t))
plt.plot(t,f3(t))
plt.grid()
plt.show()

plt.figure(figsize = (7, 2))
plt.plot(t,f4(t),'r')
plt.plot(t,f1(t)+f2(t)+f3(t), '.')
plt.grid()
plt.show()
```

Agora, observe que a função 

$$f(t)=A_0+A_1 \cos \left(\omega_0 t\right)+B_1 \operatorname{sen}\left(\omega_0 t\right) \quad \text{ em que } \quad A_1=C_1 \cos (\theta)\quad \text{e} \quad B_1=-C_1 \operatorname{sen}(\theta)$$

pode ser escrita como 

$$f(t)=A_0+C_1 \cos \left(\omega_0 t+\theta\right)$$

fazendo

$$\theta=\arctan \left(-\frac{B_1}{A_1}\right) \quad \text{e} \quad C_1=\sqrt{A_1^2+B_1^2}$$

onde, se $A_1 < 0$, soma-se $\pi$ a $\theta$.

Para escrever em função seno, basta fazer 

$$
\cos \left(\omega_0 t+\theta\right)=\operatorname{sen}\left(\omega_0 t+\theta+\frac{\pi}{2}\right.)
$$

**Ajuste por mínimos quadrados**

Podemos ajustar uma curva $f(t)=A_0+A_1 \cos \left(\omega_0 t\right)+B_1 \operatorname{sen}\left(\omega_0 t\right)$ a um conjunto discreto de $N$ dados igualmente espaçados em intervalos $\Delta t$ e com um intervalo total de $T = (N − 1)/\Delta t$ usando  mínimos quadrados ([Chapra e Canale, 2016](https://integrada.minhabiblioteca.com.br/reader/books/9788580555691/pageid/485)), fazendo:

$$A_0=\frac{\sum y}{N} , \quad A_1=\frac{2}{N} \sum y \cos(\omega_0 t) \quad \text{e} \quad B_1=\frac{2}{N} \sum y \operatorname{sen} (\omega_0 t)$$

Por exemplo, vamos considerar os dados contidos no códiogo abaixo, ajustar uma curva e plotar.

**Exemplo 1:** Ajustando uma curva $f(t)=A_0+C_1 \cos \left(\omega_0 t+\theta\right)$ a um conjunto discreto de $51$ dados igualmente espaçados.

```python
t = np.linspace(0, 2, 51)
y = np.array([ 0.29128,  0.16362,  0.63659,  0.99658,  1.32378,  1.19659,
        2.747  ,  2.45985,  2.75657,  2.22962,  2.52464,  2.75764,
        3.0182 ,  2.70341,  2.38339,  2.86054,  2.78291,  2.31757,
        2.7729 ,  1.98811,  2.09932,  1.50674,  0.88792,  0.70053,
        0.47795,  0.19647, -0.80573, -0.90089, -0.74614, -1.22724,
       -1.94578, -2.6077 , -1.87974, -2.54577, -2.62203, -2.97101,
       -2.7782 , -2.22151, -2.94329, -3.26737, -2.62968, -2.0217 ,
       -2.82572, -2.69448, -1.59839, -1.91036, -1.2359 , -0.57663,
       -1.14168, -0.57835, -0.01441])
```

```python
plt.figure(figsize = (7, 2))
plt.plot(t,y,'r.')
plt.grid()
plt.show()
```

```python
A0 = np.sum(y)/len(y)
A1 = (2/len(y))*np.sum(y*np.cos(np.pi*t))
B1 = (2/len(y))*np.sum(y*np.sin(np.pi*t))

print('A0 =',A0)
print('A1 =',A1)
print('B1 =',B1)
```

```python
plt.figure(figsize = (7, 2))
plt.plot(t,y,'r.')
plt.plot(t,A0+A1*np.cos(np.pi*t)+B1*np.sin(np.pi*t),'b')
plt.grid()
plt.show()
```

A análise anterior pode ser estendida ao modelo geral

$$f(t)=A_0 +A_1 \cos \left(\omega_0 t\right)+B_1 \operatorname{sen}\left(\omega_0 t\right)+A_2 \cos \left(2 \omega_0 t\right)+B_2 \operatorname{sen}\left(2 \omega_0 t\right) +\cdots+A_m \cos \left(m \omega_0 t\right)+B_m \operatorname{sen}\left(m \omega_0 t\right)$$

onde, para dados igualmente espaçados, os coeficientes podem ser calculados por

$$
\begin{aligned}
& A_0=\frac{\sum y}{N} \\
& \left.\begin{array}{rl}
A_j & =\frac{2}{N} \sum {y \cos \left(j \omega_0 t\right)} \\
B_j & =\frac{2}{N} \sum {y \operatorname{sen}\left(j \omega_0 t\right)}
\end{array}\right\} j=1,2, \ldots, m
\end{aligned}
$$


Essa é a abordagem utilizada na série de Fourier contínua no tempo. 

Fourier mostrou que uma função periódica arbitrária de peródo $T$ pode ser representada por uma série infinita de funções senoidais com frequências harmonicamente relacionadas. 

$$
f(t)=a_0+a_1 \cos \left(\omega_0 t\right)+b_1 \operatorname{sen}\left(\omega_0 t\right)+a_2 \cos \left(2 \omega_0 t\right)+b_2 \operatorname{sen}\left(2 \omega_0 t\right)+\cdots$$


ou, mais concisamente

$$
f(t)=a_0+\sum_{k=1}^{\infty}\left[a_k \cos \left(k \omega_0 t\right)+b_k \operatorname{sen}\left(k \omega_0 t\right)\right]$$

onde $\omega_0 = 2\pi/T$ é chamada de *frequência fundamental* e os seus múltiplos constantes $2 \omega_0$, $3\omega_0$ , etc. são chamados de $harmônicos$. 

Ou seja, a função $f(x)$ é expressa como uma combinação linear de funções base: $1, \cos \left(\omega_0 t\right), \operatorname{sen}\left(\omega_0 t\right), \cos \left(2 \omega_0 t\right), \operatorname{sen} \left(2 \omega_0 t\right),...$

Os coeficientes são calculados com 
$$a_0= \frac{1}{T} \int_{0}^{T} f(t) dt ,\quad a_k= \frac{2}{T} \int_{0}^{T} f(t) cos(k\omega_0 t) dt \quad \text{e} \quad b_k= \frac{2}{T} \int_{0}^{T} f(t) sen(k\omega_0 t) dt \quad \text{para} \quad k=1,2,3,...$$
